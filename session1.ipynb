```python
""" CORE 3.2 Coding Practice -- Script 1 

- The Iris dataset contains measurements of 150 iris flowers from three species (types): setosa, versicolor, and virginica.
- Each row = one flower (one observation).
- Each measurement is in centimeters (cm).
- “Sepal” is the outer leaf-like part of a flower (often green) that protects the bud.
  Sepal length/width measure that outer part.
- “Petal” is the colorful inner part of the flower.
  Petal length/width measure that inner part.
"""
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

""" load dataset into pandas """
# Authors: The scikit-learn developers
# SPDX-License-Identifier: BSD-3-Clause
iris = load_iris(as_frame=True)
#creates pandas dataframe
df = iris.frame 
#print(df.head())

#drop rows where the septal length is great than 5.  
#first create a boolean mask which returns true/ false for each row depending on the condition
#true means to keep the flower because the sepal length is <= 5 and false means to get rid of it
lengthbooleanfive_df = df["sepal length (cm)"] <= 5
#print(sepallengthfive_df)

lengthfive_df = df[lengthbooleanfive_df] #actually returns values instead of true/ false, dropping any that are false
print(lengthfive_df)

#outputs no. rows, no columns, can check if row count drops
#print(df.shape)
#print(lengthfive_df.shape)

 
""" Scatter plot petal v. septal length with each dot color-coded by the species of iris. """
#    - x-axis: sepal length (cm)
#    - y-axis: petal length (cm)
#    - color (c=...): species label stored in 'target' (0, 1, 2), which are labels that correspond to the three iris species.

matplotlib.pyplot.xlabel("Sepal length (cm)")
matplotlib.pyplot.ylabel("Petal length (cm)")
matplotlib.pyplot.scatter(lengthfive_df["sepal length (cm)"], 
                          lengthfive_df["petal length (cm)"],
                         c = lengthfive_df["target"])


""" acknowedgements to ChatGPT Study Mode for helping me out: https://chatgpt.com/share/69640d08-09b8-800e-93dc-ea64ffbf808c"""
```

```python
"""
CORE 3.2 Coding Practice -- Script 2

Goal of this script:
- Work with the MNIST handwritten digits dataset.
- Keep only images of the digits 3 and 8.
- Visualize one of those digit images to understand what the raw data looks like.

MNIST consists of small grayscale images of handwritten numbers (0–9).
Each image is stored as numerical pixel values rather than as a picture file.
"""

# Authors: The scikit-learn developers
# SPDX-License-Identifier: BSD-3-Clause

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd

''' load MNIST digits dataset '''
from sklearn import datasets, metrics, svm
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()

#digits.data is a 2D array:
#- each row corresponds to one handwritten digit image
#- each column corresponds to one pixel in that image
df = pd.DataFrame(digits.data)

#print(df.head())
#print(df.shape)

#contains the true digit label (i.e. the answer to what number those pixels spell out)
df["key"] = digits.target

df.head()
df.shape

''' filter dataset '''
#a boolean madk where True for rows where the digit is a 3 or 8 & apply the mask to the dataframe
mask = (df["key"] == 3) | (df["key"] == 8)
filtered_df = df[mask]

filtered_df["key"].unique() #check that only numbers 3 and 8 are there


''' printing image '''
#remove key label column so only pixel values are left
image_df = filtered_df.drop(columns=["key"])

#select row (single image), in this case, first row
row = image_df.iloc[0] 

#convert row into NumPy array of pixel values
pixels = row.values

#reshape array into grid
image = pixels.reshape(8, 8)

#show image
plt.imshow(image, cmap="gray")

'''chatgpt creds: https://chatgpt.com/share/69653118-e4ac-800e-94da-d8c7011f98d7'''
```
